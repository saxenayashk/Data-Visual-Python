```python
# Importing the libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
x=[1,7,9]
y=[1,8,11]
z=[7,5,0]

plt.plot(x,y)
plt.plot(x,z)
plt.title("Visualization of Data")
plt.xlabel("X")
plt.ylabel("Y and Z")

plt.legend(["Plot b/w X & Y" , "Plot b/w X & Z"])

plt.savefig("Visualization_of_Data1.png")
plt.show()
```

```python
# Loading the dataset
data=pd.read_csv('diabetes.csv')
```

```python
# Displaying the dataset
data
```

```python
# Scatter-plot for Glucose and BloodPressure Features
sns.scatterplot(x='Glucose',y='BloodPressure',data=data)
```

```python
# Plotting a pairplot
# Pairplot - Gains the insight wrt to each and every pair in a dataset
sns.pairplot(data)
plt.savefig("Visualization_of_Data2.png")
```

```python
# Plotting Bar-Plot: Balance Data or Not
data.Outcome.value_counts().plot(kind='bar')
```

```python
# Correlation amongst features
sns.heatmap(data.corr(),annot=True)
plt.savefig('Correlation_Data.png')
```

```python
# Distribution Plot for any Feature
# Non- Symmetric Plot - Median used for Data Imputation
sns.distplot(data.Insulin)
plt.savefig('Visualization_of_Data3.png')
```

```python
# Distrubution plot for BloodPressure
# Symmetric Plot - Mean used for Data Imputation
sns.distplot(data.BloodPressure)
plt.savefig('Visualization_of_Data4.png')
```

```python
# Box plot - detect the Outliers
fig,ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=data,ax=ax)
plt.savefig('boxPlot.png')
```
